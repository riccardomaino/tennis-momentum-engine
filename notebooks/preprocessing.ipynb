#### 4. Data Cleaning
- 4.1 Removing walkovers, unfinished matches, defaulted matches, missing score matches
- 4.2 Handle missing values
- 4.3 Convert data types

```python
print(f"Initial: {len(df)} matches in the dataset")
df = df[df["score"].notna()] # Removing matches with no STATS
print(f"Updated: {len(df)} matches left")
df = df[~df["score"].str.contains("RET", na=False)] # Removing matches with RET
print(f"Updated: {len(df)} matches left")
df = df[~df["score"].str.contains("W/O", na=False)] # Removing matches with W/O
print(f"Updated: {len(df)} matches left")
df = df[~df["score"].str.contains("DEF", na=False)] # Removing matches with DEF
print(f"Updated: {len(df)} matches left")
```

```python
# Handle missing numerical column values
num_cols = ["winner_ht", "loser_ht", "minutes"]
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[num_cols].isnull().sum()
```

```python
# Handle missing categorical column values
cat_cols_hand = ["loser_hand", "winner_hand"]
cat_cols_entry = ["winner_entry", "loser_entry"]
df[cat_cols_hand] = df[cat_cols_hand].fillna("U")
df["surface"] = df["surface"].fillna("Unknown")
df[cat_cols_entry] = df[cat_cols_entry].fillna("OK")
```

```python
# Convert data types
df["tourney_date"] = pd.to_datetime(df["tourney_date"], format="%Y%m%d")
```
