#### 4. Data Cleaning
- 4.1 Removing walkovers, unfinished matches, defaulted matches, missing score matches
- 4.2 Handle missing values and datatypes

```python
invalid_tokens = r"W/O|RET|DEF|ABD|UNF|NAN"
numeric_cols = ["w_ace", "w_df", "w_svpt", "w_1stIn", "w_1stWon", "w_2ndWon", "w_bpSaved", "w_bpFaced", "w_SvGms",
            "l_ace", "l_df", "l_svpt", "l_1stIn", "l_1stWon", "l_2ndWon", "l_bpSaved", "l_bpFaced", "l_SvGms",
            "winner_rank", "winner_rank_points", "loser_rank_points", "loser_rank",
            "winner_age", "winner_ht", "loser_age", "loser_ht", 
            "best_of"
]

# Filter out matches with no score
df = df[df["score"].notna()]
# Filter out retirements, walkovers, defaults (stats are skewed/incomplete)
df = df[~df["score"].str.contains(invalid_tokens, case=False, na=False)].copy()
# Handle missing `surface` value
df["surface"] = df["surface"].fillna("Unknown")
# Convert all the numeric columns to numeric
for col in numeric_cols:
  df[col] = pd.to_numeric(df[col], errors="coerce")
# Convert dates 
df["tourney_date"] = pd.to_datetime(df["tourney_date"], format="%Y%m%d")

df.info()
```

#### 5. Feature Engineering
For each match we compute:
  - Serve rates for winner and loser (ace %, df %, 1st serve %, etc.)
  - Breakpoints Save & Conversion rates (bp saved %, bp conversion %)
  - Return points won (= 1 - opponent serve points won)
  - Delta features (winner minus loser)
  - Context features (surface one-hot, tournament level one-hot, best-of binary feature)
  - Binary feature "came_back" and "upset"

Define also a Player Profile for each player as a new dataset.

```python
def g(prefix: str, col: str) -> str:
      return prefix+col
```

```python
# Binary Features: Comeback
cleaned_scores = df["score"].str.replace(r"\(\d+\)", "", regex=True)
first_sets = cleaned_scores.str.split().str[0]
set_split = first_sets.str.split("-", expand=True)
w_s1_games = pd.to_numeric(set_split[0], errors="coerce")
l_s1_games = pd.to_numeric(set_split[1], errors="coerce")
df["came_back"] = (w_s1_games < l_s1_games).astype(int)

df[["score", "came_back"]].head(5)
```

```python
# Binary Features: Upset
df["upset"] = np.where(df["winner_rank"] > df["loser_rank"], 1, 0)

df[["winner_rank", "loser_rank", "upset"]].head(5)
df["winner_rank"].isnull().sum()
```

```python
# Features: Winner & Loser Serve %
new_serve_features = ["ace_rate", "df_rate", "1stIn_pct", "1stWon_pct", "2ndWon_pct", "srvWon_pct"]

df["w_ace_rate"] = df["w_ace"] / df["w_svpt"]
df["w_df_rate"] = df["w_df"] / df["w_svpt"]
df["w_1stIn_pct"] = df["w_1stIn"] / df["w_svpt"]
df["w_1stWon_pct"] = df["w_1stWon"] / df["w_1stIn"]
df["w_2ndWon_pct"] = df["w_2ndWon"] / (df["w_svpt"] - df["w_1stIn"]) 
df["w_srvWon_pct"] = (df["w_1stWon"] + df["w_2ndWon"]) / df["w_svpt"]

df["l_ace_rate"] = df["l_ace"] / df["l_svpt"]
df["l_df_rate"] = df["l_df"] / df["l_svpt"]
df["l_1stIn_pct"] = df["l_1stIn"] / df["l_svpt"]
df["l_1stWon_pct"] = df["l_1stWon"] / df["l_1stIn"]
df["l_2ndWon_pct"] = df["l_2ndWon"] / (df["l_svpt"] - df["l_1stIn"]) 
df["l_srvWon_pct"] = (df["l_1stWon"] + df["l_2ndWon"]) / df["l_svpt"]

full_serve_features = [g("w_",f) for f in new_serve_features] + [g("l_", f) for f in new_serve_features]
df[full_serve_features].head(5)
# df[[g("w_", f) for f in new_serve_features]].isnull().sum()
```

```python
# Features: Breakpoints Stats %
new_breakpoints_features = ["bpSaved_pct", "bpConversion_pct"]

df["w_bpSaved_pct"] = df["w_bpSaved"] / df["w_bpFaced"]
df["l_bpSaved_pct"] = df["l_bpSaved"] / df["l_bpFaced"]
df["w_bpConversion_pct"] = 1 - df["l_bpSaved_pct"] 
df["l_bpConversion_pct"] = 1 - df["w_bpSaved_pct"]

full_breakpoints_features = [g("w_",f) for f in new_breakpoints_features] + [g("l_", f) for f in new_breakpoints_features]
df[full_breakpoints_features].head(5)
# df[[g("w_", f) for f in new_breakpoints_features]].isnull().sum()
```

```python
# Features: Return Points Won %
df["w_retWon_pct"] = 1 - df["l_srvWon_pct"] 
df["l_retWon_pct"] = 1 - df["w_srvWon_pct"]
df[["l_srvWon_pct", "w_retWon_pct", "w_srvWon_pct", "l_retWon_pct"]].head(5)
```

```python
# Feature: Ranking Gap
df["rank_gap"] = df["winner_rank"] - df["loser_rank"]
df[["winner_rank", "loser_rank", "rank_gap"]].head(5)
```

```python
# Features: Delta Features
df["delta_ace_pct"] = df["w_ace_rate"] - df["l_ace_rate"]
df["delta_1stWon_pct"] = df["w_1stWon_pct"] - df["l_1stWon_pct"]
df["delta_bpSaved_pct"] = df["w_bpSaved_pct"] - df["l_bpSaved_pct"]
df["delta_srvWon_pct"] = df["w_srvWon_pct"] - df["l_srvWon_pct"]
df["delta_retWon_pct"] = df["w_retWon_pct"] - df["l_retWon_pct"]
```

```python
# Features: Binary One-Hot Features (Context)
df = df.assign(
  surf_hard = (df["surface"] == "Hard").astype(int),
  surf_clay = (df["surface"] == "Clay").astype(int),
  surf_grass = (df["surface"] == "Grass").astype(int),
  surf_carpet = (df["surface"] == "Carpet").astype(int),
  
  is_gs = (df["tourney_level"] == "G").astype(int),
  is_masters = (df["tourney_level"] == "M").astype(int),
  is_atp = (df["tourney_level"] == "A").astype(int),
  is_finals = (df["tourney_level"] == "F").astype(int),
  is_davis = (df["tourney_level"] == "D").astype(int),
  
  best_of_5 = (df["best_of"] == 5).astype(int)
)
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=["w_ace_rate", "l_ace_rate", "came_back"])
df = df.drop(columns=["surface", "draw_size", "tourney_level", "match_num", "best_of", "round",
                        "winner_seed", "winner_entry", "winner_ioc", "winner_rank_points",
                        "loser_seed", "loser_entry", "loser_ioc", "loser_rank_points"])
df = df.reset_index(drop=True)
```

##### Define Player Profiles

```python
SERVE_COLS = ["ace", "df", "svpt", "1stIn", "1stWon", "2ndWon", "bpSaved", "bpFaced"] # prefix: "w_"=winner, "l_"=loser
w_serve_cols = [g("w_", col) for col in SERVE_COLS]
l_serve_cols = [g("l_", col) for col in SERVE_COLS]

# Create a Winners & Losers slices
winners = df[["winner_name"] + w_serve_cols + ["came_back"]].copy()
# winners.columns = ["player", "ace", "svpt", "came_back"]
winners["is_winner"] = 1
winners.head(5)
```

```python
losers = df[["loser_name", "l_ace", "l_svpt"]].copy()
losers.columns = ["player", "ace", "svpt"]
losers["came_back"] = 0 # Loser cannot comeback
losers["is_winner"] = 0
losers.head(5)
```

```python
player_stats = pd.concat([winners, losers], axis=0)
player_stats.head(5)
```
